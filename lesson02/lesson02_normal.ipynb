```python
__author__ = 'Алексей Сидорюк'

# Задача-1:
# Дан список, заполненный произвольными целыми числами, получите новый список,
# элементами которого будут квадратные корни элементов исходного списка,
# но только если результаты извлечения корня не имеют десятичной части и
# если такой корень вообще можно извлечь
# Пример: Дано: [2, -5, 8, 9, -25, 25, 4]   Результат: [3, 5, 2]

import math

list1 = [7, -5, 8, 49, -0, 25, 4, 'test', 100]
list2 = []
for idx, x in enumerate(list1):
    if (type(x) is int):
        if (x > 0):
            if math.sqrt(x).is_integer():
                list2.append(int(math.sqrt(x)))
print(list2)
```

```python
# Задача-2: Дана дата в формате dd.mm.yyyy, например: 02.11.2013.
# Ваша задача вывести дату в текстовом виде, например: второе ноября 2013 года.
# Склонением пренебречь (2000 года, 2010 года)

date_str = '11.13.1981'
list1 = date_str.split('.')

rec_date = {'dd': {'01': 'первое',
                   '02': 'второе',
                   '03': 'третье',
                   '04': 'четвертое',
                   '05': 'пятое',
                   '06': 'шестое',
                   '07': 'седьмое',
                   '08': 'восьмое',
                   '09': 'девятое',
                   '20': 'десятое',
                   '11': 'одиннадцатое',
                   '12': 'двенадцатое',
                   '13': 'тринадцатое',
                   '14': 'четырнадцатое',
                   '15': 'пятнадцатое',
                   '16': 'шестнадцатое',
                   '17': 'семнадцатое',
                   '18': 'восемнадцатое',
                   '19': 'девятнадцатое',
                   '20': 'двадцатое',
                   '21': 'двадцать первое',
                   '22': 'двадцать второе',
                   '23': 'двадцать третье',
                   '24': 'двадцать четвертое',
                   '25': 'двадцать пятое',
                   '26': 'двадцать шестое',
                   '27': 'двадцать седьмое',
                   '28': 'двадцать восьмое',
                   '29': 'двадцать девятое',
                   '30': 'тридцатое',
                   '31': 'тридцать первое'},
            'mm': {'01':'января',
                   '02':'февраля',
                   '03':'марта',
                   '04':'апреля',
                   '05':'мая',
                   '06':'июня',
                   '07':'июля',
                   '08':'августа',
                   '09':'сентября',
                   '10':'октября',
                   '11':'ноября',
                   '12':'декабря'}}

for key in rec_date['dd']:
    if list1[0] == key:
        list1[0] = rec_date['dd'][key]
for key in rec_date['mm']:
    if list1[1] == key:
        list1[1] = rec_date['mm'][key]        
list1[2] = list1[2] + ' года'

date_str = ' '.join(list1)

print(date_str)
```

```python
# Задача-3: Напишите алгоритм, заполняющий список произвольными целыми числами
# в диапазоне от -100 до 100. В списке должно быть n - элементов.
# Подсказка:
# для получения случайного числа используйте функцию randint() модуля random

import random

list1 = []
n = input('Введите целое положительное число: ')
if n.isdigit():
    if int(n) > 0:
        for i in range(1, int(n)):
            list1.append(random.randint(-100, 100))
else:
    print('Число должно быть целым и положительным.')
print(list1)
```
